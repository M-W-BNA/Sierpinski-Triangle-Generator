```python
import turtle
import time
import numpy as np
from copy import deepcopy
```

```python
def sierpinski_recursive(size, initial, counter):
    remove_list = []
    print(mid_points_queue)
    for pos in mid_points_queue:
        if initial == False:
            tri.penup()
        tri.goto(pos[0], pos[1])
        tri.setheading(0) 
        if counter == 1:
            edge_coords.append(np.round(tri.position(),2))
        
        
        if pos[-1] == "=": 
            tri.setheading(300)
            edge_coords.append(np.round(tri.position(),2))
            
        if pos[-1] == "-":
            tri.setheading(180)
            tri.left(60)
            tri.forward(size)
            tri.setheading(300)
            edge_coords.append(np.round(tri.position(),2))

        if pos[-1] == "+":
            tri.setheading(0)
            tri.right(60)
            tri.forward(size)
            tri.setheading(300)
            edge_coords.append(np.round(tri.position(),2))

        tri.pendown()

        for i in range(3):
            if pos[-1] in ["+","-","="]:
                tri.left(120)
                tri.forward(size)
                if i < 2:
                    edge_coords.append(np.round(tri.position(),2))
                 
        if mid_points_queue[-1] == pos:
            if initial == False:
                tri.penup()
            tri.goto(mid_points_queue[0][0], mid_points_queue[0][1])
        remove_list.append(pos)

    initial = False
    for val in remove_list:
        if val in mid_points_queue:
            mid_points_queue.remove(val)

    return_edge = deepcopy(edge_coords[0:len(edge_coords)-1:3])
    return_counter = 0

    for i in range(3**counter):
        if (i+1) % 3 == 0:
            line = tuple(np.add(np.array(edge_coords[0]), np.array(return_edge[return_counter])).tolist())
            second_edge_coord = return_edge[return_counter]
            return_counter+=1
       
        else:
            line = tuple(np.add(np.array(edge_coords[0]), np.array(edge_coords[1])).tolist())
            second_edge_coord = edge_coords[1]

        mid_point = tuple(np.round((ti/2),2) for ti in line)

        grad_mid_point = list(mid_point)

        if edge_coords[0][0] < second_edge_coord[0] and edge_coords[0][1] < second_edge_coord[1]:
            grad_mid_point.append("+")
        elif edge_coords[0][1] == second_edge_coord[1]:
            grad_mid_point.append("=")
        else:
            grad_mid_point.append("-")
 
        mid_point = tuple(grad_mid_point)

        mid_points_queue.append(mid_point)
        del edge_coords[0]

    if size < 25:
        turtle.done()
    counter += 1

    sierpinski_recursive(size/2, initial, counter) 
```

```python
wn = turtle.Screen()
tri = turtle.Turtle()
tri.clear()
tri.speed(10)
tri.hideturtle()
edge_coords = [(0,0)]
positions = [(0,0)]
size = 400
mid_points_queue = []

for i in range(3):
    tri.forward(size)
    tri.left(120)
    positions.append(tri.position())

cx = sum(p[0] for p in positions) / len(positions)
cy = sum(p[1] for p in positions) / len(positions)
offset_x = -cx
offset_y = -cy
tri.clear()
tri.penup()


centered_positions = [(x + offset_x, y + offset_y) for (x, y) in positions]

tri.goto(centered_positions[0])
tri.pendown()

for point in centered_positions[1:] + [centered_positions[0]]:
    tri.goto(point)

first = True

for i in range(3):
    edge_coords.append(tri.position())
    line = tuple(np.add(np.array(centered_positions[i]), np.array(centered_positions[(i+1)%3])).tolist())
    mid_point = tuple(ti/2 for ti in line)
    mid_points_queue.append(mid_point)
edge_coords.clear()

size = size/2
initial = True
counter = 1
sierpinski_recursive(size, initial, counter)
```
